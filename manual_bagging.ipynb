prepare the data for bootstrapping

```python
data_df=pd.concat([X,y],axis=1)
print(data_df.head(10))
```

bootstrapping

```python
def bootstrap_v(dataset,n_samples):
    samples=[]
    for j in range(n_samples):
        start=100
        end=len(dataset)-1
        sample=[]
        l=random.randint(start,end)
        for _ in range(0,l):
            index = random.randint(0,end)
            sample.append(dataset.iloc[index])
        len_org=len(sample)-1 # length of rows with no duplicat
        print(len_org)
        while(len(sample)<len(dataset)):
            k=random.randint(0,len_org)
            sample.append(sample[k])
        samples.append(sample)
    return samples
```

split as train and test

```python
def Samples(data):
    trainSamples=[]
    testSamples=[]
    print("Total number of data samples:",len(data))
    train_num=int(input("Enter the number of training samples:"))
    i=0
    while(i<train_num):
       index = random.randrange(len(data))
       trainSamples.append(data.iloc[index])
       i=i+1
    test_num=int(input("Enter the number of test samples "))
    i=0
    while(i<test_num):
       index = random.randrange(len(data))
       testSamples.append(data.iloc[index])
       i=i+1
    train_df=pd.DataFrame(trainSamples,columns=data.columns,index=[i for i in range(0,len(trainSamples))])
    test_df=pd.DataFrame(testSamples,columns=data.columns,index=[i for i in range(0,len(testSamples))])
    return train_df,test_df
```

main program

```python
train_df,test_df=Samples(data_df)
```

```python
num_samples=int(input("Enter the number of data samples"))
samples_data=bootstrap_v(train_df,num_samples)
```

```python
df1=pd.DataFrame(samples_data[0],columns=data_df.columns,index=[i for i in range(0,len(train_df))])
df1.head(5)
```

```python
op_list=[]
test_feat=test_df.copy()
del test_feat["Diagnosis"]
for i in range(0,num_samples):
  print("=====================DATA",i,"======================================")
  df_temp=pd.DataFrame(samples_data[i],columns=data_df.columns,index=[i for i in range(0,len(train_df))])
  y=df_temp.Diagnosis
  X=df_temp.copy()
  del X["Diagnosis"]
  clf=DecisionTreeClassifier(
            criterion = "entropy")
  clf.fit(X,y)
  pred=clf.predict(test_feat)
  pred_list=list(pred)
  op_list.append(pred_list)
  pred_df=pd.DataFrame(pred_list,columns=['predicted_class'])
  temp_df=pd.concat([test_df,pred_df],axis=1)
  print(temp_df.head(10))
  print("======================================================================")
```

```python
final_pred=[]
vote_list=[]
for n in range(0,len(test_df)):
    votes=0
    temp=[]
    for j in range(0,num_samples):
        temp.append(op_list[j][n])
    final_pred.append(mode(temp))
    vote_list.append(temp.count(final_pred[n]))
```

```python
pred_f_df=pd.DataFrame(final_pred,columns=['Predicted'])
voting_df=pd.DataFrame(vote_list,columns=['Voting'])
pred_df_final=pd.concat([pred_f_df,voting_df],axis=1)
final_df=pd.concat([test_df,pred_df_final],axis=1)
print("Sample result")
print(final_df.head(10))
```
